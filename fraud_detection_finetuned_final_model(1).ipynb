```python
from transformers import DebertaV2Tokenizer, DebertaV2ForSequenceClassification
import torch

tokenizer = DebertaV2Tokenizer.from_pretrained("microsoft/deberta-v3-base")

def tokenize(batch):
    tokenized_inputs = tokenizer(
        batch["conversations"],
        padding="max_length",
        truncation=True,
        max_length=256,
        return_tensors=None
    )

    if len(batch["conversations"]) > 0:
        print(f"Sample text length: {len(batch['conversations'][0])}")
        print(f"Sample tokenized length: {len(tokenized_inputs['input_ids'][0])}")

    return {
        "input_ids": tokenized_inputs["input_ids"],
        "attention_mask": tokenized_inputs["attention_mask"],
        "labels": batch["label"]
    }

dataset_encoded = dataset.map(
    tokenize,
    batched=True,
    batch_size=32
)

dataset_encoded.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "labels"]
)

print("Available splits:", dataset_encoded.keys())

first_item = dataset_encoded["train"][0]

print("Input shape:", first_item["input_ids"].shape)
print("Attention mask shape:", first_item["attention_mask"].shape)
```

```python
import tensorflow as tf
from transformers import TFAutoModel

class DeBERTaForClassification(tf.keras.Model):
    def __init__(self, model_name, num_classes, dropout_rate=0.3):
        super().__init__()
        self.deberta = TFAutoModel.from_pretrained(model_name)
        self.dropout = tf.keras.layers.Dropout(dropout_rate)  # Add Dropout
        self.fc = tf.keras.layers.Dense(num_classes, activation='softmax')

    def call(self, inputs, training=False):
        x = self.deberta(inputs)[0][:, 0, :]  # CLS token
        x = self.dropout(x, training=training)  # Apply dropout only during training
        return self.fc(x)

```

```python
classifier = DeBERTaForClassification("microsoft/deberta-v3-base", num_classes=2)

from transformers import AdamWeightDecay

classifier.compile(
    optimizer=AdamWeightDecay(learning_rate=2e-5, weight_decay_rate=0.01),  # Add weight decay
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_true = []  # True labels from the test dataset
y_pred = []  # Model predictions

for batch in test_dataset:
    batch_inputs, batch_labels = batch
    preds = classifier.predict(batch_inputs)
    preds = tf.argmax(preds, axis=1).numpy()

    y_true.extend(batch_labels.numpy())
    y_pred.extend(preds)

# Generate Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Normal", "Fraud"], yticklabels=["Normal", "Fraud"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

```

```python
from sklearn.metrics import roc_curve, auc

y_true = []
y_scores = []

for batch in test_dataset:
    batch_inputs, batch_labels = batch
    preds = classifier.predict(batch_inputs)

    # Extract fraud probabilities
    probs = tf.nn.softmax(preds).numpy()[:, 1]

    y_true.extend(batch_labels.numpy())
    y_scores.extend(probs)

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Plot
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```
